# Density

We call a product for which a 
country has developed RCA, an occupied product (O), and one for which it has not an 
unoccupied product (U). When we compare two time points there are 4 possible 
transitions (U->U,U->O,O->U,O->O), and in our case, we are concerned with the second 
one which takes unoccupied products to occupied ones. Additionally, we call a product
undergoing this particular transition: transition products.

We now ask: are transition products closer to occupied products than to unoccupied ones? If this is significantly the 
case, it would be evidence supporting that countries perform structural transformations by jumping from occupied products to nearby ones.
To proof this we need to define some quantitiesn. First, we measured the average proximity of a new potential product $j$ to a country’s current productive structure, which we call
density and define as

$$\omega^k_j=\frac{\sum_i x_i\phi_{ij}}{\sum_i\phi_{ij}}$$

where $\omega^k_j$ is the density around good $j$ given the export basket of the $k$th country and $x_i = 1$ if $RCA_{ki} > 1$ and 0 otherwise. A high density value means that the $k$th country has many developed products surrounding the $j$th product.

```python
# Sum in the up side of the density equation
upsum_omega_2004 = np.matmul(M_2004,Prox_2004)
upsum_omega_2009 = np.matmul(M_2009,Prox_2009)
upsum_omega_2014 = np.matmul(M_2014,Prox_2014)
upsum_omega_2019 = np.matmul(M_2019,Prox_2019)
omega_2004 = np.zeros(upsum_omega_2004.shape)
omega_2009 = np.zeros(upsum_omega_2009.shape)
omega_2014 = np.zeros(upsum_omega_2014.shape)
omega_2019 = np.zeros(upsum_omega_2019.shape)


#division by lower sum
for j in range(0,Prox_2004.shape[1]):
    omega_2004[:,j] = upsum_omega_2004[:,j]/Prox_2004[:,j].sum()
    
for j in range(0,Prox_2009.shape[1]):
    omega_2009[:,j] = upsum_omega_2009[:,j]/Prox_2009[:,j].sum()
    
for j in range(0,Prox_2014.shape[1]):
    omega_2014[:,j] = upsum_omega_2014[:,j]/Prox_2014[:,j].sum()
    
for j in range(0,Prox_2019.shape[1]):
    omega_2019[:,j] = upsum_omega_2019[:,j]/Prox_2019[:,j].sum()
```

```python
#Build datafremes
omega_2004_df = pd.DataFrame(columns=M_Columns_2004)
omega_2004_df['Estado'] = M_Rows_2004

for i,est in enumerate(M_Rows_2004):
    for j,prod in enumerate(M_Columns_2004):
        omega_2004_df.loc[omega_2004_df['Estado']==est,prod] = omega_2004[i,j]
        
omega_2009_df = pd.DataFrame(columns=M_Columns_2009)
omega_2009_df['Estado'] = M_Rows_2004

for i,est in enumerate(M_Rows_2004):
    for j,prod in enumerate(M_Columns_2009):
        omega_2009_df.loc[omega_2009_df['Estado']==est,prod] = omega_2009[i,j]
        
omega_2014_df = pd.DataFrame(columns=M_Columns_2014)
omega_2014_df['Estado'] = M_Rows_2004

for i,est in enumerate(M_Rows_2004):
    for j,prod in enumerate(M_Columns_2014):
        omega_2014_df.loc[omega_2014_df['Estado']==est,prod] = omega_2014[i,j]
        
omega_2019_df = pd.DataFrame(columns=M_Columns_2019)
omega_2019_df['Estado'] = M_Rows_2004

for i,est in enumerate(M_Rows_2004):
    for j,prod in enumerate(M_Columns_2019):
        omega_2019_df.loc[omega_2019_df['Estado']==est,prod] = omega_2019[i,j]
```

## Discovery Factor

At the single product level, we considered the ratio between the average density of all countries in which the jth product was a transition product and the average density of all countries in which the $j$th product was not developed. Formally, we define the “discovery factor” $H_j$ as

$$H_j=\frac{\sum_{k=1}^T\omega_j^k/T}{\sum_{k=T+1}^N\omega_j^k/(N-T)}$$

where $T$ is the number of countries in which the $j$th good was a transition product and $N$ is the total number of countries.

```python
N=32
Discovery_04_09 = pd.DataFrame(columns=common_prod_04_09,index=[0])
for prod in common_prod_04_09:
    trans=np.array(Transition_04_09[prod].values)
    no_trans=np.array(underdeveloped_04_09[prod].values)
    T=sum(trans)
    dens = np.array(omega_2004_df[prod].values)
    if np.dot(dens,no_trans)!=0 and T!=0:
        H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
    else:
        H_j = 0
    
    Discovery_04_09[prod] = H_j
    
Discovery_04_14 = pd.DataFrame(columns=common_prod_04_14,index=[0])
for prod in common_prod_04_14:
    trans=np.array(Transition_04_14[prod].values)
    no_trans=np.array(underdeveloped_04_14[prod].values)
    T=sum(trans)
    dens = np.array(omega_2004_df[prod].values)
    if np.dot(dens,no_trans)!=0 and T!=0:
        H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
    else:
        H_j = 0
    
    Discovery_04_14[prod] = H_j
    
Discovery_04_19 = pd.DataFrame(columns=common_prod_04_19,index=[0])
for prod in common_prod_04_19:
    trans=np.array(Transition_04_19[prod].values)
    no_trans=np.array(underdeveloped_04_19[prod].values)
    T=sum(trans)
    dens = np.array(omega_2004_df[prod].values)
    if np.dot(dens,no_trans)!=0 and T!=0:
        H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
    else:
        H_j = 0
    
    Discovery_04_19[prod] = H_j
    
Discovery_09_14 = pd.DataFrame(columns=common_prod_09_14,index=[0])
for prod in common_prod_09_14:
    trans=np.array(Transition_09_14[prod].values)
    no_trans=np.array(underdeveloped_09_14[prod].values)
    T=sum(trans)
    dens = np.array(omega_2009_df[prod].values)
    if np.dot(dens,no_trans)!=0 and T!=0:
        H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
    else:
        H_j = 0
    
    Discovery_09_14[prod] = H_j
    
Discovery_09_19 = pd.DataFrame(columns=common_prod_09_19,index=[0])
for prod in common_prod_09_19:
    trans=np.array(Transition_09_19[prod].values)
    no_trans=np.array(underdeveloped_09_19[prod].values)
    T=sum(trans)
    dens = np.array(omega_2009_df[prod].values)
    if np.dot(dens,no_trans)!=0 and T!=0:
        H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
    else:
        H_j = 0
    
    Discovery_09_19[prod] = H_j
    
Discovery_14_19 = pd.DataFrame(columns=common_prod_14_19,index=[0])
for prod in common_prod_14_19:
    trans=np.array(Transition_14_19[prod].values)
    no_trans=np.array(underdeveloped_14_19[prod].values)
    T=sum(trans)
    dens = np.array(omega_2014_df[prod].values)
    if np.dot(dens,no_trans)!=0 and T!=0:
        H_j = (N-T)/T * np.dot(dens,trans)/ np.dot(dens,no_trans)
    else:
        H_j = 0
    
    Discovery_14_19[prod] = H_j
```
