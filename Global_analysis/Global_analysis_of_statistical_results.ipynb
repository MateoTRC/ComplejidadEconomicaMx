```python
import pandas as pd
import numpy as np
```

```python
TEST_Data_cae_ue_folder='../Data_cae_ue/ANOVA_Kurskal_test/'
TEST_Data_edo_cae_folder='../Data_edo_cae/ANOVA_Kurskal_test/'
TEST_Data_pot_cae_folder='../Data_pot/ANOVA_Kurskal_test/'
```

```python
df_Data_cae_ue_04_09=pd.read_csv(TEST_Data_cae_ue_folder+'ANOVA_Kurskal_test_04_09.csv',index_col=0)
df_Data_cae_ue_09_14=pd.read_csv(TEST_Data_cae_ue_folder+'ANOVA_Kurskal_test_09_14.csv',index_col=0)
df_Data_cae_ue_14_19=pd.read_csv(TEST_Data_cae_ue_folder+'ANOVA_Kurskal_test_14_19.csv',index_col=0)

df_Data_edo_cae_04_09=pd.read_csv(TEST_Data_edo_cae_folder+'ANOVA_Kurskal_test_04_09.csv',index_col=0)
df_Data_edo_cae_09_14=pd.read_csv(TEST_Data_edo_cae_folder+'ANOVA_Kurskal_test_09_14.csv',index_col=0)
df_Data_edo_cae_14_19=pd.read_csv(TEST_Data_edo_cae_folder+'ANOVA_Kurskal_test_14_19.csv',index_col=0)

df_Data_pot_cae_04_09=pd.read_csv(TEST_Data_pot_cae_folder+'ANOVA_Kurskal_test_04_09.csv',index_col=0)
df_Data_pot_cae_09_14=pd.read_csv(TEST_Data_pot_cae_folder+'ANOVA_Kurskal_test_09_14.csv',index_col=0)
df_Data_pot_cae_14_19=pd.read_csv(TEST_Data_pot_cae_folder+'ANOVA_Kurskal_test_14_19.csv',index_col=0)
```

```python
df_Data_cae_ue_04_09_0_ANOVA=df_Data_cae_ue_04_09[df_Data_cae_ue_04_09['ANOVA_flag']==0]
df_Data_cae_ue_04_09_0_ANOVA['Data_set']='Data_cae_ue'
df_Data_cae_ue_09_14_0_ANOVA=df_Data_cae_ue_09_14[df_Data_cae_ue_09_14['ANOVA_flag']==0]
df_Data_cae_ue_09_14_0_ANOVA['Data_set']='Data_cae_ue'
df_Data_cae_ue_14_19_0_ANOVA=df_Data_cae_ue_14_19[df_Data_cae_ue_14_19['ANOVA_flag']==0]
df_Data_cae_ue_14_19_0_ANOVA['Data_set']='Data_cae_ue'

df_Data_cae_ue_ALL_0_ANOVA=pd.concat([df_Data_cae_ue_04_09_0_ANOVA,df_Data_cae_ue_09_14_0_ANOVA,df_Data_cae_ue_14_19_0_ANOVA])

import os

directory_name = "CSVs"

# Check if the directory exists
if not os.path.exists(directory_name):
    # Create the directory
    os.makedirs(directory_name)
    print(f"Directory '{directory_name}' created successfully.")
else:
    print(f"Directory '{directory_name}' already exists.")


df_Data_cae_ue_ALL_0_ANOVA.to_csv('CSVs/Data_cae_ue_ALL_0_ANOVA.csv')

df_Data_cae_ue_ALL_0_ANOVA

```

```python
df_common_states_Data_cae_ue_ANOVA = df_Data_cae_ue_04_09_0_ANOVA.merge(df_Data_cae_ue_09_14_0_ANOVA, on='Estado', how='inner')
df_common_states_Data_cae_ue_ANOVA = df_common_states_Data_cae_ue_ANOVA.merge(df_Data_cae_ue_14_19_0_ANOVA, on='Estado', how='inner')

df_common_states_Data_cae_ue_ANOVA=df_common_states_Data_cae_ue_ANOVA[['Estado','ANOVA_flag','Kurskal_flag','Data_set']]

df_common_states_Data_cae_ue_ANOVA.to_csv('CSVs/common_states_Data_cae_ue_ANOVA.csv')

df_common_states_Data_cae_ue_ANOVA
```
